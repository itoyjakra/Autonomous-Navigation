```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
%matplotlib inline
import pandas as pd
```

```python
im = mpimg.imread('/ARTIFACTS/Autonomous-Navigation/DATA/CARLA/out-1013A/episode_0000/CameraRGB/000894.png')
fig, ax = plt.subplots(1, 1, figsize=[16, 10])
ax.imshow(im)
```

```python
index = 0
drive_hist = []
with open("/ARTIFACTS/Autonomous-Navigation/DATA/CARLA/out-1013A/screen.out-1013A") as f:
    for line in f:
        #if index > 100:
        #    break
        if "Vehicle" in line:
            
            #print(line)
            s1 = line.split('),')[0]
            co = s1.split('at (')[-1]
            x = float(co.split(',')[0])
            y = float(co.split(',')[-1])
            
            n = int(line.split('/')[-1])
            #print('----', co, x, y, n)
            params = {'x': x, 'y': y, 'n': n}
            
        if "steer" in line:
            #print(line)
            steer = float(line.split('steer:')[-1])
            #print(steer)
            params['steer'] = steer
            drive_hist.append(params)
            index +=1 
```

```python
df_dhist = pd.DataFrame(drive_hist)
```
